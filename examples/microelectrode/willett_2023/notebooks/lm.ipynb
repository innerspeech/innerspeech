```python
from datasets import load_dataset

dataset = load_dataset("deven367/babylm-100M-switchboard")
```

```python
# join all train and valid
train = dataset['train']['text']
```

```python
# join all train in the list
# train_text = ' '.join(train)
```

```python
# write to a file
with open('train.txt', 'w') as f:
    for train_text in train:
        # clean all punctuation
        train_text = train_text.lower()
        train_text = train_text.replace('.', '')
        train_text = train_text.replace('?', '')
        train_text = train_text.replace('!', '')
        train_text = train_text.replace(',', '')
        train_text = train_text.replace('-', '')

        f.write(train_text + '\n')
```

```python
# train sentence piece model based on the train text

import sentencepiece as spm
vocab_size = 512
spm.SentencePieceTrainer.train(input='train.txt', model_prefix=f'pretrained/m{vocab_size}', vocab_size=vocab_size)
```
